Consider the simple function R(β) = sin(β) + β/10.
(a) Draw a graph of this function over the range β ∈ [−6, 6].
(b) What is the derivative of this function?
(c) Given β0 = 2.3, run gradient descent to find a local minimum
of R(β) using a learning rate of ρ = 0.1. Show each of β0
, β1
, . . .
in your plot, as well as the final answer.
(d) Repeat with β0 = 1.4.

```python
# Define the function and its derivative
def R(beta):
    return np.sin(beta) + beta/10

def R_derivative(beta):
    return np.cos(beta) + 1/10

# (a) Plot the function
beta = np.linspace(-6, 6, 1000)
y = R(beta)

plt.figure(figsize=(6, 3))
plt.plot(beta, y, 'b-', label='R(β)')
plt.grid(True)
plt.xlabel('β')
plt.ylabel('R(β)')
plt.title('R(β) = sin(β) + β/10')
plt.legend()
plt.show()

# (c) Gradient descent from β₀ = 2.3
def gradient_descent(start_beta, learning_rate, n_iterations=100, tolerance=1e-6):
    beta_history = [start_beta]
    beta = start_beta

    for i in range(n_iterations):
        gradient = R_derivative(beta)
        new_beta = beta - learning_rate * gradient

        # Check for convergence
        if abs(new_beta - beta) < tolerance:
            break

        beta = new_beta
        beta_history.append(beta)

    return np.array(beta_history)

# Run gradient descent for both starting points
beta_history_1 = gradient_descent(2.3, 0.1)
beta_history_2 = gradient_descent(1.4, 0.1)

# Plot results
plt.figure(figsize=(6, 3))
plt.plot(beta, y, 'b-', label='R(β)')

# Plot first gradient descent path
plt.plot(beta_history_1, R(beta_history_1), 'ro-', label='GD from β₀=2.3')
plt.plot(beta_history_1[0], R(beta_history_1[0]), 'go', label='Start β₀=2.3')
plt.plot(beta_history_1[-1], R(beta_history_1[-1]), 'ko', label='End β₀=2.3')

# Plot second gradient descent path
plt.plot(beta_history_2, R(beta_history_2), 'mo-', label='GD from β₀=1.4')
plt.plot(beta_history_2[0], R(beta_history_2[0]), 'yo', label='Start β₀=1.4')
plt.plot(beta_history_2[-1], R(beta_history_2[-1]), 'ko', label='End β₀=1.4')

plt.grid(True)
plt.xlabel('β')
plt.ylabel('R(β)')
plt.title('Gradient Descent Paths')
plt.show()

# Print final values
print(f"Final β from starting point 2.3: {beta_history_1[-1]:.4f}")
print(f"Final β from starting point 1.4: {beta_history_2[-1]:.4f}")
print(f"Number of iterations (2.3): {len(beta_history_1)}")
print(f"Number of iterations (1.4): {len(beta_history_2)}")
```
